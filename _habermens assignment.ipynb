# PDF & CDF

```python
# pdf of operation Age
```

```python
sns.FacetGrid(haberman, hue='Surv_status', size=8)\
 .map(sns.distplot,"Age")\
 .add_legend()
plt.show()
```

in above plot age range from 30-75 the status of survival and death is same 

```python
#pdf of Auxillary nodes

sns.FacetGrid(haberman, hue='Surv_status', size=8)\
 .map(sns.distplot,"Axil_Nodes")\
 .add_legend()
plt.show()
```

it has been observed that people survive long if they have less auxillary nodes

# CDF

```python


haberman_Survival_years = haberman.loc[haberman["Surv_status"] == 1]
counts,bin_edges = np.histogram(haberman_Survival_years["Age"], bins = 10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

CDF gives information of patient age, such as 60 % of patients had less than 10 nodes

```python

haberman_Survival_years = haberman.loc[haberman["Surv_status"] == 2]
counts,bin_edges = np.histogram(haberman_Survival_years["Axil_Nodes"], bins = 10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
